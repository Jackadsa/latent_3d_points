```python


def vae_template(bneck_size, bneck_post_mlp=False, encoder_args = None, decoder_args = None):
    ''' Single class experiments.
    '''
#     if n_pc_points != 2048:
#         raise ValueError()

    encoder = variational_encoder_with_convs_and_symmetry
    decoder = decoder_with_fc_only

    n_input = [100, 3]
    n_output = [500,2]
    
    if encoder_args is None:
        encoder_args = {'n_filters': [256, 256, 256],
                        'filter_sizes': [1],
                        'strides': [1],
                        'b_norm': True,
                        'verbose': True,
                        'symmetry': tf.reduce_mean,
                        'fully_connected_layers': [256,256,256],
                        'latent_size':bneck_size,
                        'bnorm_momentum':0.99,
                        'bnorm_clipping':{'rmin':1,'rmax':1,'dmax':0},
                        'EFN': False
                        }
    if decoder_args is None:
        decoder_args = {'b_norm': True,
                        'b_norm_finish': True,
                        'verbose': True,
                        'layer_sizes': [512,1024,1024,1000],
                        'reshape_last_connected': [500,2],
                        'bnorm_momentum':0.99,
                        'bnorm_clipping':{'rmin':1,'rmax':1,'dmax':0}
                        }

    if bneck_post_mlp:
        encoder_args['n_filters'].pop()
        decoder_args['layer_sizes'][0] = bneck_size

    return encoder, decoder, encoder_args, decoder_args

encoder, decoder, enc_args, dec_args = vae_template(bneck_size)
```
